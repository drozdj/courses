## 5.7 Scalar-matrix multiplication

```python
display(Image('data/19072024_214632.png'))
import math

delta = 15
matrix = np.array(([[1,2],[3,4]]))
viz = np.zeros((2, 2), dtype='U16')

np.multiply(delta, matrix)

# a bit of manual visualizing doesn't hurt anybody
viz[0, 0] = f'{delta} * {matrix[0, 0]} = {delta * matrix[0, 0]}'
viz[0, 1] = f'{delta} * {matrix[0, 1]} = {delta * matrix[0, 1]}'
viz[1, 0] = f'{delta} * {matrix[1, 0]} = {delta * matrix[1, 0]}'
viz[1, 1] = f'{delta} * {matrix[1, 1]} = {delta * matrix[1, 1]}'
display(viz)


d_matrix = matrix * delta
d_matrix
```

```python
gcd = reduce(gcd, np.array(d_matrix).flatten()) # 15
print(f'GCD: {gcd}')

display(np.divide(d_matrix, gcd))
```
